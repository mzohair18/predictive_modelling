1.)Dealing with missing values

```python
#transform [city-mpg] column to [city-lpg]
df['city-mpg']=235/df['city-mpg']
```

```python
df.rename(columns={'city-mpg':'city-L/100kms'},inplace=True)
```

```python
df["price"].dtypes
```

# finding Missing values in the column Price

```python
df['price'].isnull().sum()
```

```python
#filling null/missing values in every column which has null values
mean_value=df['price'].mean() 
norm=df['normalized-losses'].mean()
hp=df['horsepower'].mean()
es=df['engine-size'].mean()
hmpg=df['highway-mpg'].mean()



df['price'].fillna(value=mean_value, inplace=True)
df['normalized-losses'].fillna(value=norm,inplace=True)
df['horsepower'].fillna(value=hp, inplace=True)
df['engine-size'].fillna(value=es,inplace=True)
df['highway-mpg'].fillna(value=hmpg,inplace=True)
```

```python
#converting the price column to int as this is the target variable we need during modelling
df['price']=df['price'].astype(int)
```

# Data normalization (making the data consistent)

When features are in very large and different ranges, we apply max,min-max and z-score normalization

The goal of normalization is to transform the data into a common scale without distorting the differences in the ranges of values.

```python
df['length'] = df['length'] / df['length'].max()
```

```python
df['length']
```

# Bining ( grouping data into small bins)

It involves dividing a continuous variable into a set of smaller intervals or bins and replacing the original values with the corresponding bin labels.

For eg you can bin age into 0-5,6-10,11-15...

```python
import numpy as np
```

```python
bins=np.linspace(min(df['price']),max(df['price']),4)
```

```python
bins
```

```python
group_names=["Low","Medium","High"]
```

```python
#create a variables and store bins in it
df['priced-bins']=pd.cut(df['price'],bins,labels=group_names,include_lowest=True)
```

'Multiple Linear Regression' refers to more than one independant variables to make prediction
